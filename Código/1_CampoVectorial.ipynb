
En esta primera versión, sólo definimos el campo vectorial para que los robots alcancen la trayectoria. Todo esto en $R^2$, por lo que $n = 2$ ($n \equiv dimensión$). En otro código trataremos el caso en $R^3$

Implementamos las bibliotecas pertinentes

```python
import numpy as np
import matplotlib.pyplot as plt
```

Definimos los parámetros y funciones de la simulación

```python
n = 2
t_total = 100
paso = 0.1
t = np.linspace(0, paso, t_total)
dx = 2
dy = 2

def f(t):
    fx = np.cos(t)
    fy = np.sin(t)
    
    return np.array([fx, fy])

#Está mal definido esto
def df(t,h=1e-5):
    Df = (f(t+h)-f(t-h))/(2*h)

    return np.array([Df[0], Df[1]])
```

Definimos el campo vectorial a $t = 0$, junto con la curva que recorren los robots.

```python
tc = np.linspace(0,2*np.pi,360)
dibujo = f(tc)
curva = f(0)
Deriv = df(0)
```

Definimos el campo vectorial

```python
x0 = np.linspace(-2, 2, 10)
y0 = np.linspace(-2, 2, 10)

kx = 1
ky = 1

vx = []
vy = []

for i in range(len(x0)):
    vx.append(((-1)**n*Deriv[0]) - kx*(x0[i] - curva[0]))
    vy.append(((-1)**n*Deriv[1]) - ky*(y0[i] - curva[1]))

Vx,Vy = np.meshgrid(vx,vy)
X0,Y0 = np.meshgrid(x0,y0)
```

Representamos el campo vectorial

```python
plt.quiver(X0,Y0,Vx,Vy)
plt.plot(dibujo[0], dibujo[1], color='red', linewidth=2)
plt.show
```
