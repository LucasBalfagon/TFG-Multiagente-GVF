
En esta primera versión, sólo definimos el campo vectorial para que los robots alcancen la trayectoria. Todo esto en $R^2$, por lo que $n = 2$ ($n \equiv dimensión$). En otro código trataremos el caso en $R^3$

Implementamos las bibliotecas pertinentes

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact
import matplotlib.legend_handler as handler
```

Definimos los parámetros y funciones de la simulación

```python
n = 2

def f(t):
    fx = t*np.cos(t*2.5)/4
    fy = t*np.sin(t)/4
    
    return np.array([fx, fy])


def df(t,h=1e-5):
    Df = (f(t+h)-f(t-h))/(2*h)

    return np.array([Df[0], Df[1]])
```

Definimos la curva a la que deben tender los robots.

```python
tc = np.linspace(0,3*np.pi,400)
dibujo = f(tc)
```

Definimos los parámetros necesarios para representar el campo vectorial

```python
x0 = np.linspace(-2, 2, 9)
y0 = np.linspace(-2, 2, 9)

X0,Y0 = np.meshgrid(x0,y0)

kx = 1
ky = 1
```

Representamos el campo vectorial, junto con el slider. El punto rojo es a donde tienden los robots, antes de aplicar el control a sus distancias.

```python
def slider(w=0):
    curva = f(w)
    Deriv = df(w)
    
    vx = []
    vy = []

    for i in range(len(x0)):
        vx.append(((-1)**n*Deriv[0]) - kx*(x0[i] - curva[0]))
        vy.append(((-1)**n*Deriv[1]) - ky*(y0[i] - curva[1]))

    Vx,Vy = np.meshgrid(vx,vy)

    plt.quiver(X0,Y0,Vx,Vy)
    plt.plot(curva[0],curva[1], color='red', marker='o',label='Posición deseada')
    plt.plot(dibujo[0], dibujo[1], color='red', linewidth=2,label='Trayectoria final')
    plt.legend()
    plt.show()

interact(slider, w=(0, 9, 0.001))
```
